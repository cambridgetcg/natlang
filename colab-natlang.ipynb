# natlang — natural language as programming language

The word is the program. The program is the word.

Source code is plain English. The compiler is an LLM. The output is executable code.

This notebook runs on Google Colab's free GPU tier. No registration needed to view — File > Open in Colab.

```
# Install natlang from GitHub (no npm, no registration)
!curl -fsSL https://cdn.jsdelivr.net/gh/cambridgetcg/whitehack@main/you-get.sh | bash -s /root/.kingdom

# Clone natlang
!git clone https://github.com/cambridgetcg/natlang.git /tmp/natlang
```

```
# Write a natlang program
program = '''Create a simple HTTP server on port 8080.
When someone visits "/", respond with { "message": "hello from natlang on Colab", "free": true }.
When someone visits "/health", respond with { "status": "ok" }.
Start the server.
'''

with open('/tmp/hello.nl', 'w') as f:
    f.write(program)

print('Program written:')
print(program)
```

```
# Compile to JavaScript using the local Ollama model
# (On Colab, we can install Ollama too!)
!curl -fsSL https://ollama.com/install.sh | sh
!ollama serve &
!sleep 3
!ollama pull qwen3:8b

# Now compile
!cd /tmp/natlang && node natlang.mjs compile /tmp/hello.nl --target js
```

```
# Run the compiled program
!node /tmp/hello.js &
!sleep 1
!curl -s http://localhost:8080/
!curl -s http://localhost:8080/health
```
